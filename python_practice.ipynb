### Using `input()` in Python

The `input()` function in Python is used to get user input from the console. It reads a line from the input (usually `sys.stdin`), converts it to a string, and returns that string. You can optionally pass a string argument to `input()`, which will be displayed as a prompt to the user.

#### Basic Usage: Getting String Input

The simplest way to use `input()` is to just capture text from the user.

```python
# Get a simple string input
user_name = input("What is your name? ")
print(f"Hello, {user_name}!")
```

#### Getting Numeric Input (Integers and Floats)

It's very important to remember that `input()` *always* returns a string. If you need to perform mathematical operations or comparisons that require numbers, you must convert the input string to an integer (`int()`) or a float (`float()`).

```python
# Get integer input
age_str = input("How old are you? ")
age = int(age_str) # Convert the string to an integer
print(f"You are {age} years old.")
print(f"Next year you will be {age + 1}!")

# Get float input
height_str = input("How tall are you in meters? ")
height = float(height_str) # Convert the string to a float
print(f"Your height is {height} meters.")
```

#### Combining Prompt and Conversion

You can often combine the prompt and the type conversion in a single line for cleaner code.

```python
# Combined input and conversion for an integer
favorite_number = int(input("What is your favorite number? "))
print(f"Your favorite number is {favorite_number}.")

# Combined input and conversion for a float
price = float(input("Enter a price: "))
print(f"The price is ${price:.2f}")
```
